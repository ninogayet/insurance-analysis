```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
import matplotlib.pyplot as plt

#Data Preparation
#Extracting and computing insurance datas
data = pd.read_csv("insurance.csv")
data.head()
data.info()
data.describe()
```

Output 1 - Distribution of charges

```python
plt.hist(data.charges)
plt.show()
```

Output 2 - Smokers vs Non-Smokers

```python
data_s = data[data['smoker'] == 'yes']['charges']
data_ns = data[data['smoker'] == 'no']['charges']

plt.figure(figsize=(12, 5))
plt.bar('Smokers', data_s.mean(), color='red')
plt.bar('Non-Smokers', data_ns.mean(), color='green')
plt.title('Smokers and Non-Smokers Charges comparation')
plt.ylabel('Charges (USD)')
plt.show()
```

Output 3 - Age vs Charges

```python
plt.figure(figsize=(12, 5))
plt.scatter(data['age'], data['charges'])
plt.title('Age and Charges correlation')
plt.xlabel('Age')
plt.ylabel('Charges (USD)')
plt.show()
```

Output 4 - Age vs Charges colored by smoker status

```python
colors = data['smoker'].map({'yes': 'red', 'no': 'blue'})
plt.scatter(data['age'], data['charges'], c=colors)
```

```python
#Let's put all graphs together !

import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3, ax4) = plt.subplots(4, 1, figsize=(12, 20))

# Output 1 — Distribution of charges
ax1.hist(data['charges'])
ax1.set_title('Distribution of Insurance Charges')
ax1.set_xlabel('Charges (USD)')
ax1.set_ylabel('Count')

# Output 2 — Smokers vs Non-Smokers
data_s = data[data['smoker'] == 'yes']['charges']
data_ns = data[data['smoker'] == 'no']['charges']
ax2.bar('Smokers', data_s.mean(), color='red')
ax2.bar('Non-Smokers', data_ns.mean(), color='green')
ax2.set_title('Average Charges — Smokers vs Non-Smokers')
ax2.set_ylabel('Average Charges (USD)')

# Output 3 — Age vs Charges
ax3.scatter(data['age'], data['charges'])
ax3.set_title('Age and Charges Correlation')
ax3.set_xlabel('Age')
ax3.set_ylabel('Charges (USD)')

# Output 4 — Age vs Charges colored by smoker status
colors = data['smoker'].map({'yes': 'red', 'no': 'blue'})
ax4.scatter(data['age'], data['charges'], c=colors)
ax4.set_title('Age and Charges — Red: Smokers, Blue: Non-Smokers')
ax4.set_xlabel('Age')
ax4.set_ylabel('Charges (USD)')

plt.tight_layout()
plt.show()
```
